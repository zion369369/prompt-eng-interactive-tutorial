## 📖 The Google Prompting Framework
According to the official **Google Workspace Prompting Guide 101**, an effective prompt is built on four core pillars:
1. **Persona**: Who the model should act as (e.g., *You are a senior editor*).
2. **Task**: What the model needs to do (e.g., *Draft a newsletter summary*). Always start with action verbs.
3. **Context**: Background details and references (e.g., *Using this product launch timeline*).
4. **Format**: The desired output layout and constraints (e.g., *In a markdown table, under 200 words*).

While not every prompt requires all four elements, using them consistently will drastically improve the relevance and accuracy of Gemini's outputs.

## ⚙️ Administrative Example: Washington DC Offsite Agenda
Let's look at an official example from the Google Prompting Guide for planning a team offsite. Notice how it maps to the four elements:
- **Persona**: *Executive administrator to a team director*
- **Context**: *Newly formed team of content, digital, and product marketers; gathering for a three-day offsite in Washington, DC*
- **Task**: *Plan activities for each day that include team bonding and strategic work; create a sample agenda*
- **Format**: *Markdown table layout*


```python
# Define the model using gemini-3.5-flash
model = genai.GenerativeModel(
    model_name='gemini-3.5-flash',
    system_instruction="You are an executive administrator to a team director. The team consists of content marketers, digital marketers, and product marketers."
)

prompt = """
We are gathering for the first time at a three-day offsite in Washington, DC. 
Plan activities for each day that include team bonding activities and time for deeper strategic work. 
Create a sample agenda formatted as a markdown table.
"""

# To run this, uncomment below:
# response = model.generate_content(prompt)
# print(response.text)
```

## 🏋️ Interactive Exercise: Customer Service Response
Now it's your turn! Write a prompt to handle a customer complaint about damaged goods. Your prompt must incorporate:
1. **Persona**: A customer service representative.
2. **Task**: Craft an empathetic email response.
3. **Context**: The customer ordered headphones that arrived damaged; they want expedited shipping on the replacement (which isn't typically allowed).
4. **Format**: A paragraph acknowledging frustration followed by exactly three bullet points with potential resolutions.

Write your prompt as a string in the variable `user_prompt` below.


```python
# Write your prompt here:
user_prompt = """
[Replace this text with your prompt incorporating Persona, Task, Context, and Format]
"""
```

```python
# Run this cell to validate your prompt structure!
def validate_framework(p):
    if "[Replace this text" in p or not p.strip():
        print("❌ Please replace the template text with your own prompt first!")
        return
    
    score = 0
    criteria = []
    
    # Check Persona
    persona_keywords = ["representative", "agent", "specialist", "counselor", "support", "rep", "customer service"]
    if any(kw in p.lower() for kw in persona_keywords):
        score += 1
        criteria.append("✅ Persona included (e.g. customer service representative)")
    else:
        criteria.append("⚠️ Missing clear persona definition (e.g. 'I am a customer service agent')")
        
    # Check Task
    task_keywords = ["write", "draft", "craft", "email", "respond", "reply", "send"]
    if any(kw in p.lower() for kw in task_keywords):
        score += 1
        criteria.append("✅ Task included (e.g. draft an email response)")
    else:
        criteria.append("⚠️ Missing clear action/task instruction (e.g. 'write an empathetic reply')")
        
    # Check Context
    context_keywords = ["damage", "headphone", "expedite", "ship", "deliver", "broken"]
    if any(kw in p.lower() for kw in context_keywords):
        score += 1
        criteria.append("✅ Context included (e.g. damaged headphones and shipping request)")
    else:
        criteria.append("⚠️ Missing background context details (e.g. damaged headphones details)")
        
    # Check Format
    format_keywords = ["bullet", "resolution", "paragraph", "frustration", "options", "points"]
    if any(kw in p.lower() for kw in format_keywords):
        score += 1
        criteria.append("✅ Format constraints included (e.g. paragraph and three bullet points)")
    else:
        criteria.append("⚠️ Missing format structure requirements (e.g. 'three bullet points')")
        
    print(f"Validation Score: {score}/4")
    for check in criteria:
        print(check)
        
    if score == 4:
        print("\n🎉 Excellent work! Your prompt is a Google-standard power prompt!")
    else:
        print("\n💡 Tip: Refine your prompt to address the warning points above!")

validate_framework(user_prompt)
```
