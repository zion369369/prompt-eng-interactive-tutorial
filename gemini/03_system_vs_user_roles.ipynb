# ♊ Gemini Chapter 3: System Instructions vs. Content Roles

Gemini models distinguish between high-level **system instructions** and conversational **contents**.

---

## 🛠️ The Architecture of Roles
1. **System Instruction (`system_instruction`)**: Defines the global state, target persona, behavior guidelines, and output constraints. It is passed *outside* the message history and remains constant.
2. **User Message (`user`)**: The input data, context files, or user request.
3. **Model Message (`model`)**: The response generated by Gemini.

Using system instructions prevents the model from forgetting its rules over long multi-turn conversations.


```python
import google.generativeai as genai

# Configuring a system instruction
model = genai.GenerativeModel(
    model_name='gemini-3.5-flash',
    system_instruction="""
    You are a communications advisor who only writes official company announcements.
    Always use a formal, confident tone. Keep memos under 150 words.
    """
)
```

## 🏋️ Interactive Exercise: Customizing System Roles
You need to build a system instruction for a **PR Spokesperson Mock Q&A assistant**.
The assistant should prepare a corporate executive for interviews. Write a python script below where:
1. The `system_instruction` establishes a PR manager persona who generates mock interview questions (mix of easy and hard questions) and reviews the executive's answers.
2. The user prompt supplies details about a newly launched product (e.g. `@Product Launch`).

Implement the setup below.


```python
# Define your system instruction and model here:
system_instruction = """
[Define the Persona, Guidelines, and Tone for the PR Q&A Assistant here]
"""

# Define the user prompt here (supplying the context details):
user_content = """
[Provide the product launch details or questions to evaluate]
"""
```

```python
# Validator to check system instruction config
def validate_roles(sys_inst, usr_cont):
    if "[Define" in sys_inst or "[Provide" in usr_cont:
        print("❌ Please fill in both the system instruction and user content variables!")
        return
        
    score = 0
    criteria = []
    
    # Check system instruction depth
    sys_words = len(sys_inst.split())
    if sys_words >= 15:
        score += 1
        criteria.append(f"✅ System instruction is descriptive ({sys_words} words)")
    else:
        criteria.append("⚠️ System instruction is very short. Provide more rules and persona details.")
        
    # Check persona/spokesperson indicator in system instruction
    keywords = ["pr", "public relations", "spokesperson", "interview", "question", "qa", "coach", "prep"]
    if any(k in sys_inst.lower() for k in keywords):
        score += 1
        criteria.append("✅ System instruction targets PR coach/questioning context")
    else:
        criteria.append("⚠️ System instruction should explicitly define a PR coach or interviewer role.")
        
    print(f"Validation score: {score}/2")
    for c in criteria:
        print(c)
        
    if score == 2:
        print("\n🎉 System and content roles correctly designed!")
    else:
        print("\n💡 Revise your roles using the suggestions above.")

validate_roles(system_instruction, user_content)
```
