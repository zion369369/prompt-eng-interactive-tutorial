## 1. Core Anatomy of a Gemini Prompt

Gemini APIs are structured differently from older completion models. A request typically comprises:
*   **System Instructions (`system_instruction`)**: Defines the persona, constraints, and rules for the model.
*   **Contents (`contents`)**: The history of turns containing the actual prompt content, supports multi-modal files (text, images, audio, video).
*   **Generation Config (`generation_config`)**: Controls parameters like `temperature`, `top_p`, `response_mime_type`, and `response_schema` (for structured JSON outputs).
