## 1. Core Anatomy of a Gemini Prompt

Gemini APIs are structured differently from older completion models. A request typically comprises:
*   **System Instructions (`system_instruction`)**: Defines the persona, constraints, and rules for the model.
*   **Contents (`contents`)**: The history of turns containing the actual prompt content, supports multi-modal files (text, images, audio, video).
*   **Generation Config (`generation_config`)**: Controls parameters like `temperature`, `top_p`, `response_mime_type`, and `response_schema` (for structured JSON outputs).

## 2. Using System Instructions (Persona, Context, Task, Format)

System instructions are highly respected by Gemini. In this example from the Google 101 guide, we establish an executive administrator persona to plan a team offsite agenda.

```python
# Example: Planning a team offsite using the Persona-Task-Context-Format framework
model = genai.GenerativeModel(
    model_name='gemini-3.5-flash',
    system_instruction="""
    You are an executive administrator to a team director. 
    The team consists of content marketers, digital marketers, and product marketers.
    Create high-quality, professional schedules matching these constraints.
    """
)

prompt = """
We are gathering for the first time at a three-day offsite in Washington, DC. 
Plan activities for each day that include team bonding activities and time for deeper strategic work. 
Create a sample agenda formatted as a markdown table.
"""

# response = model.generate_content(prompt)
# print(response.text)
```
