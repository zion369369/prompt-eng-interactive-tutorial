## 📦 GenerationConfig & response_schema
You can configure Gemini to return a JSON string and enforce a specific Python class structure using `response_mime_type="application/json"` and `response_schema`.

This relies on `typing.TypedDict` or `pydantic.BaseModel` to outline the exact keys, types, and descriptions of the keys.


```python
import google.generativeai as genai
import typing_extensions as typing

# Schema for expense item tracking
class ExpenseItem(typing.TypedDict):
    date: str
    expense_type: str  # Meal, Transport, Lodging, Entertainment
    vendor_name: str
    description: str
    amount_usd: float

model = genai.GenerativeModel('gemini-3.5-flash')
# When generating content, you configure it like so:
# response = model.generate_content(
#     "Create a sample list of 2 expenses for Seattle travel.",
#     generation_config=genai.types.GenerationConfig(
#         response_mime_type="application/json",
#         response_schema=ExpenseItem
#     )
# )
```

Create a structured output schema for customer satisfaction surveys as referenced in the Google 101 guide.

Define a `TypedDict` subclass named `CustomerSurvey` that includes:
1. `agent_effectiveness`: string representing if the call was resolved effectively.
2. `concern_addressed`: boolean representing if the primary concern was solved.
3. `recommendation_score`: integer from 1 to 10.
4. `customer_feedback`: string with open text feedback.

Write your schema definition below.


```python
import typing_extensions as typing

# Define your TypedDict schema here:
# class CustomerSurvey(typing.TypedDict):
#     # YOUR CODE HERE
#     ...
```

```python
# Validator to check class definition
import typing_extensions as typing

def check_schema():
    # Check if CustomerSurvey is defined in globals
    if 'CustomerSurvey' not in globals() and 'CustomerSurvey' not in locals():
        print("❌ CustomerSurvey class is not defined. Please define the TypedDict subclass.")
        return
        
    survey_cls = globals().get('CustomerSurvey', None)
    if not issubclass(survey_cls, dict):
        print("❌ CustomerSurvey must inherit from typing.TypedDict!")
        return
        
    annotations = survey_cls.__annotations__
    score = 0
    criteria = []
    
    fields = [
        ('agent_effectiveness', str),
        ('concern_addressed', bool),
        ('recommendation_score', int),
        ('customer_feedback', str)
    ]
    
    for field, ftype in fields:
        if field in annotations:
            if annotations[field] == ftype:
                score += 1
                criteria.append(f"✅ Field '{field}' matches type {ftype.__name__}")
            else:
                criteria.append(f"⚠️ Field '{field}' has type {annotations[field].__name__}, expected {ftype.__name__}")
        else:
            criteria.append(f"❌ Missing field '{field}'")
            
    print(f"Schema Match Score: {score}/4")
    for c in criteria:
        print(c)
        
    if score == 4:
        print("\n🎉 Perfect! Your structured output schema is fully ready for Gemini's response_schema config!")
    else:
        print("\n💡 Modify the TypedDict fields to match the instructions.")

try:
    check_schema()
except Exception as e:
    print(f"❌ Validation error: {e}")
```
