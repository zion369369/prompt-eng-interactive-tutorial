# ♊ Gemini Chapter 5: Advanced Prompt Chaining & Multi-Modal Inputs

Gemini is natively multi-modal, meaning it can process text, images, audio, and video directly within a single context window.

Additionally, complex tasks are often best solved using **Prompt Chaining**—where you pass the output of one model request into a subsequent request to perform multiple logical steps.

---

## 🔗 Prompt Chaining
Instead of asking Gemini to write a perfect document in one go, split it up:
1. **Step 1**: Summarize raw inbox transcripts or document notes.
2. **Step 2**: Generate high-level milestones or action items from that summary.
3. **Step 3**: Draft a formal stakeholder notification email based on the action items.

This minimizes hallucinations and provides modular checkpoints.
