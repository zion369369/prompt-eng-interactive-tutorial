## 💡 The 21-Word Rule
Official research from Google shows that the most successful prompts average **21 words** or more, incorporating relevant context. Conversely, typical user prompts average less than nine words, which leads to generic, low-quality results.

By expanding your prompts with specific details, context, and constraints, you turn a simple prompt into a **power prompt**.

Let's look at how to expand prompts using Google's **6 quick tips for success**:
1. **Use Natural Language**: Write in full sentences as if speaking to a human colleague.
2. **Be Specific**: Tell the model exactly what to do (summarize, write, change tone).
3. **Avoid Complexity**: State requests in brief, specific language without unnecessary jargon.
4. **Make it a Conversation**: Use follow-up prompts to refine output.
5. **Reference Your Context**: Supply relevant context or data directly in the prompt.
6. **Use Gemini as an Editor**: Let Gemini upgrade your prompts: *'Make this a power prompt: [original prompt text]'*

## 🏋️ Interactive Exercise: Prompt Expansion
Take the following weak prompt:
> Write a budget tracker.

Rewrite it into a **power prompt** that has **at least 21 words** and specifies:
1. **Persona**: A project manager or startup founder.
2. **Context**: Tracking expenses for business travel.
3. **Format**: Specific column headers (e.g. date, expense type, vendor name, amount) formatted as a table.

Write your expanded prompt below in `power_prompt`.


```python
# Write your expanded prompt (must be 21+ words!):
power_prompt = """
[Replace this text with your expanded prompt of at least 21 words]
"""
```

```python
# Run this cell to validate the 21-Word Rule and prompt content!
def validate_21_words(p):
    if "[Replace this text" in p or not p.strip():
        print("❌ Please replace the template text first!")
        return
    
    words = p.split()
    word_count = len(words)
    print(f"Your prompt contains {word_count} words.")
    
    score = 0
    criteria = []
    
    if word_count >= 21:
        score += 1
        criteria.append("✅ Rule Met: Prompt has 21+ words!")
    else:
        criteria.append(f"❌ Rule Failed: Prompt is only {word_count} words. Add more context or format specifics to reach 21+ words.")
        
    # Check context keywords
    context_kw = ["budget", "tracker", "expense", "travel", "trip", "column", "table", "date", "vendor"]
    found_kw = [kw for kw in context_kw if kw in p.lower()]
    if len(found_kw) >= 3:
        score += 1
        criteria.append(f"✅ Good depth: Found keywords: {found_kw}")
    else:
        criteria.append("⚠️ Add more specific keywords (e.g. expense, travel, table, columns)")
        
    print(f"Validation score: {score}/2")
    for c in criteria:
        print(c)
        
    if score == 2:
        print("\n🎉 Great job! You successfully crafted a high-context 21+ word power prompt!")
    else:
        print("\n💡 Try expanding the detail to hit both criteria.")

validate_21_words(power_prompt)
```
