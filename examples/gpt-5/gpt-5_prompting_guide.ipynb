## 📖 1. The ChatGPT 4-Part Framework
To get high-quality responses from ChatGPT, structure your prompts using Google's **Persona, Task, Context, and Format** framework:

1.  **Persona**: Who ChatGPT should act as (*"Act as a senior PR manager..."*).
2.  **Task**: What you want ChatGPT to do (*"Draft a press release..."*). Always start with action verbs.
3.  **Context**: Background details and files (*"Here is our product launch roadmap: [text or upload PDF]..."*).
4.  **Format**: How the output should look (*"Format as a markdown table under 300 words..."*).

### ⚙️ Example Comparison
- **Weak Prompt (3 words)**:
  > Write an agenda.
- **Framework Power Prompt (65 words)**:
  > **[Persona]** Act as an executive administrator to a team director. 
  > **[Context]** Our newly formed team consists of content marketers and product marketers. We are gathering for a three-day offsite in Washington, DC. 
  > **[Task]** Plan activities for each day that include team bonding and strategic planning. 
  > **[Format]** Create a sample agenda formatted as a markdown table.


## 💡 2. The 21-Word Rule & Conversation Tips
Google's research shows that the most successful prompts average **21 words** or more with rich context, whereas typical user prompts are under 9 words.

### ⚡ 5 Tips for ChatGPT Conversations:
1.  **Use Natural Language**: Speak to ChatGPT as you would to a smart human assistant.
2.  **Make it a Conversation**: Don't expect perfection in one go. Iterate and ask for adjustments.
3.  **Upload Documents**: Use the paperclip icon in ChatGPT to upload PDF reports, Excel budgets, or whiteboard photos for direct context.
4.  **Ask for Feedback**: Ask ChatGPT: *"What questions do you have for me that would help you write a better offsite agenda?"*
5.  **Use ChatGPT as a Prompt Editor**: Start your query with: *"Make this a power prompt: [your basic prompt text]"*


## 🔄 3. Interactive Prompting (Clarification Loop)
One of the best ways to prompt ChatGPT is to instruct it to ask you questions *before* it generates the final output. This ensures it has all the necessary context.

### 📝 Try copying this template into ChatGPT:
```text
You are a professional copywriter. I want you to help me write a landing page script for our new Chrome extension.
Before you start writing, ask me 3 clarifying questions one by one about our target audience, unique value proposition, and brand voice.
Once I answer, you will write the script.
```

## 🏋️ Interactive Practice: Write Your Own Prompt
Use the cell below to draft a power prompt for ChatGPT that combines **Persona, Task, Context, and Format**.
Ensure it is **at least 21 words** long.


````python
# Write your prompt string inside the variable below:
my_chatgpt_prompt = """
[Replace this with your expanded prompt of at least 21 words]
"""
````

````python
# Run this cell to validate if your prompt meets the ChatGPT 21-Word and Framework rules!
def check_chatgpt_prompt(p):
    if "[Replace" in p or not p.strip():
        print("❌ Please replace the template text with your prompt first!")
        return
        
    word_count = len(p.split())
    print(f"Word count: {word_count}")
    
    score = 0
    criteria = []
    
    if word_count >= 21:
        score += 1
        criteria.append("✅ Rule Met: Prompt is 21+ words (great context depth)")
    else:
        criteria.append(f"❌ Rule Failed: Prompt is only {word_count} words. Add more details to reach 21+ words.")
        
    # Check for Persona indicators
    persona_kw = ["act as", "you are", "role", "persona", "assistant", "representative", "manager", "founder", "writer"]
    if any(k in p.lower() for k in persona_kw):
        score += 1
        criteria.append("✅ Persona established (e.g. 'You are a professional...')")
    else:
        criteria.append("⚠️ No clear Persona found (e.g. 'You are a copywriter')")
        
    # Check for Format indicators
    format_kw = ["format", "table", "bullet", "paragraph", "list", "markdown", "words", "length"]
    if any(k in p.lower() for k in format_kw):
        score += 1
        criteria.append("✅ Output Format specified (e.g. 'markdown table', 'bullet points')")
    else:
        criteria.append("⚠️ No clear Output Format specified (e.g. 'limit to three paragraphs')")
        
    print(f"Validation score: {score}/3")
    for c in criteria:
        print(c)
        
    if score == 3:
        print("\n🎉 Perfect! This prompt is structured optimally for ChatGPT (GPT-5).")
    else:
        print("\n💡 Try expanding the prompt to satisfy all guidelines.")

check_chatgpt_prompt(my_chatgpt_prompt)
````
