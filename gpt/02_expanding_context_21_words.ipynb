## 💡 The 21-Word Rule in GPT
To get top-tier results from reasoning models like `gpt-5.5-pro` and helpers like `gpt-5.4-mini`, avoid short prompts.
Ensure your user instructions are **at least 21 words**, combining context and constraints.

Let's practice transforming a generic prompt into an expanded, high-quality prompt.

## 🏋️ Interactive Exercise: Expand a travel prompt
Take this weak prompt:
> Plan my manager's business trip.

Expand it into a prompt of **at least 21 words** containing:
1. **Persona**: Executive assistant.
2. **Context**: Two-day business trip to Chicago, manager staying at Hyatt Regency.
3. **Format**: Table of breakfast/dinner options within walking distance.

Write your expanded prompt below.


```python
# Write your expanded prompt (must be 21+ words!):
expanded_prompt = """
[Replace this text with your expanded prompt of at least 21 words]
"""
```

```python
# Run validation
def validate_gpt_expansion(p):
    if "[Replace" in p or not p.strip():
        print("❌ Please replace the template text first!")
        return
        
    word_count = len(p.split())
    print(f"Word count: {word_count}")
    
    score = 0
    criteria = []
    
    if word_count >= 21:
        score += 1
        criteria.append("✅ Rule Met: Prompt is 21+ words")
    else:
        criteria.append(f"❌ Rule Failed: Prompt is only {word_count} words. Add more details.")
        
    kw = ["trip", "chicago", "hyatt", "walking", "distance", "table", "breakfast", "dinner"]
    found = [k for k in kw if k in p.lower()]
    if len(found) >= 3:
        score += 1
        criteria.append(f"✅ High detail depth (keywords found: {found})")
    else:
        criteria.append("⚠️ Add more location/itinerary specifics to the prompt context.")
        
    print(f"Validation score: {score}/2")
    for c in criteria:
        print(c)
        
    if score == 2:
        print("\n🎉 Great work! This prompt gives GPT the necessary context to generate precise answers.")
    else:
        print("\n💡 Try expanding the detail to hit both criteria.")

validate_gpt_expansion(expanded_prompt)
```
