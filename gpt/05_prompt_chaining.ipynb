## 🔗 The Chaining Pattern
In a chaining pipeline, you run the first model completion, capture the response, and inject it as part of the user input for the second completion step.

This is highly effective for tasks like content drafting, translation reviews, and multi-step math tasks.

## 🏋️ Interactive Exercise: Chain a Translation & Review Pipeline
Let's design a two-step chaining program to translate a customer feedback message and then evaluate its sentiment:
1. **Step 1**: Translate input text into English.
2. **Step 2**: Analyze the sentiment of the English translation (Positive, Neutral, Negative).

Specify the prompts you would use for these two steps below.


```python
# Define Step 1 translation prompt template
translation_prompt_template = """
[Write a prompt that translates this input text into English: {input_text}]
"""

# Define Step 2 sentiment analysis prompt template
sentiment_prompt_template = """
[Write a prompt that analyzes the sentiment of this text: {translated_text}]
"""
```

```python
# Validator to check chaining variables
def validate_gpt_chain(p1, p2):
    if "[Write" in p1 or "[Write" in p2:
        print("❌ Please replace the template text first!")
        return
        
    score = 0
    criteria = []
    
    if "translate" in p1.lower() and "english" in p1.lower():
        score += 1
        criteria.append("✅ Step 1 correctly targets English translation")
    else:
        criteria.append("⚠️ Step 1 should explicitly specify translation to English")
        
    if "sentiment" in p2.lower() or any(w in p2.lower() for w in ["positive", "neutral", "negative"]):
        score += 1
        criteria.append("✅ Step 2 correctly targets sentiment evaluation")
    else:
        criteria.append("⚠️ Step 2 should analyze sentiment (positive/neutral/negative)")
        
    print(f"Validation score: {score}/2")
    for c in criteria:
        print(c)
        
    if score == 2:
        print("\n🎉 Chaining parameters correctly aligned!")
    else:
        print("\n💡 Adjust your prompt templates according to the recommendations.")

validate_gpt_chain(translation_prompt_template, sentiment_prompt_template)
```
