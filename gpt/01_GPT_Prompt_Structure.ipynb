## 1. The Chat Completions API Message Schema

OpenAI models represent prompts as a list of message objects, each representing a turn in the dialogue:
*   **System Message (`system`)**: Directs the assistant's behavior, tone, style, and rules.
*   **User Message (`user`)**: Represents the instructions or prompts sent by the user.
*   **Assistant Message (`assistant`)**: Simulates the model's responses, useful for few-shot learning.
