## 🎭 Message Roles Explained
*   **System Message (`system`)**: Persistent controller. Sets overall constraints, assistant tone, and target persona.
*   **User Message (`user`)**: The current query or context inputs.
*   **Assistant Message (`assistant`)**: Pre-populates the model's history. Extremely useful for **Few-Shot Prompting**, where you show the model examples of how it should behave.

Let's see how few-shot prompting uses the assistant role.


```python
# Example of few-shot message format
messages = [
    {"role": "system", "content": "You write corporate slogans emphasizing reliability and history."},
    {"role": "user", "content": "Company: Acme Widgets, founded in 1910"},
    {"role": "assistant", "content": "Acme Widgets: A Century of Unmatched Reliability."},
    {"role": "user", "content": "Company: Apex Solutions, founded in 1995"}
]
```

## 🏋️ Interactive Exercise: Spokesperson Prep Q&A
Create a few-shot message structure for a PR spokesperson prep assistant.
Provide a system instruction, one example exchange (user asking for a mock question, assistant responding with an example of an easy/hard question), and a final user query asking for a question about a new product announcement.

Write your structured messages list below.


```python
# Define your list of message objects
few_shot_messages = [
    # 1. System message defining PR role
    {"role": "system", "content": "[Your system prompt here]"},
    
    # 2. User example prompt
    {"role": "user", "content": "[User example]"},
    
    # 3. Assistant example response (showing correct format/tone)
    {"role": "assistant", "content": "[Assistant response example]"},
    
    # 4. Final User prompt
    {"role": "user", "content": "[Final User query]"}
]
```

```python
# Validator to check messages list
def validate_messages(msgs):
    if len(msgs) < 4:
        print("❌ Please provide at least 4 message turns for a proper few-shot dialogue!")
        return
        
    roles = [m.get("role") for m in msgs]
    contents = [m.get("content") for m in msgs]
    
    if any("[" in c for c in contents):
        print("❌ Please replace all placeholder brackets first!")
        return
        
    score = 0
    criteria = []
    
    if roles[0] == "system":
        score += 1
        criteria.append("✅ Message 1 is the 'system' rule maker")
    else:
        criteria.append("❌ Message 1 must have the role 'system'")
        
    if roles[1] == "user" and roles[2] == "assistant":
        score += 1
        criteria.append("✅ Dialogue includes user example followed by assistant response")
    else:
        criteria.append("❌ Dialogue must show a user/assistant example pair")
        
    print(f"Validation score: {score}/2")
    for c in criteria:
        print(c)
        
    if score == 2:
        print("\n🎉 Excellent! Few-shot prompting using assistant messages is the most effective way to align GPT outputs.")
    else:
        print("\n💡 Review the message sequence roles.")

validate_messages(few_shot_messages)
```
