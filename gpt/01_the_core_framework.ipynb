## 📖 Adapting the 4-Part Framework to Chat Completions
Google's 4-part framework maps directly to OpenAI message roles:
1. **Persona** -> Best placed in the `system` message to establish permanent tone and behavior rules.
2. **Context** -> Placed in the `system` or `user` message to provide background.
3. **Task** -> Placed in the `user` message as the active instruction.
4. **Format** -> Placed in the `system` or `user` message to constrain output shape.

## 🏋️ Interactive Exercise: Administrative offsite planner
Now write a prompt for a three-day offsite planner in Washington, DC using the OpenAI API mapping:
1. **System Message**: Persona of an executive administrator.
2. **User Message**: Task (plan 3 days of bonding & strategy), Context (new team of content, digital, and product marketers), and Format (markdown table agenda).

Write your prompts in the variables `system_message` and `user_message` below.


```python
system_message = """
[Define the Persona here]
"""

user_message = """
[Define the Task, Context, and Format here]
"""
```

```python
# Run this cell to validate your message objects!
def validate_gpt_framework(sys_msg, usr_msg):
    if "[Define" in sys_msg or "[Define" in usr_msg:
        print("❌ Please fill in the variables first!")
        return
        
    score = 0
    criteria = []
    
    # Check system persona
    sys_kw = ["admin", "assistant", "administrator", "secretary", "coordinator"]
    if any(k in sys_msg.lower() for k in sys_kw):
        score += 1
        criteria.append("✅ System Persona defined correctly")
    else:
        criteria.append("⚠️ System message should set an administrative persona")
        
    # Check user context/format
    user_kw = ["offsite", "washington", "bonding", "strategy", "table", "agenda", "markdown"]
    found = [k for k in user_kw if k in usr_msg.lower()]
    if len(found) >= 3:
        score += 1
        criteria.append(f"✅ User prompt contains sufficient context and constraints (keywords: {found})")
    else:
        criteria.append("⚠️ User prompt lacks key details (offsite, washington, table, agenda)")
        
    print(f"Validation score: {score}/2")
    for c in criteria:
        print(c)
        
    if score == 2:
        print("\n🎉 Perfect! You've mapped the framework into OpenAI's message layout!")
    else:
        print("\n💡 Refine your prompts to clear the warnings.")

validate_gpt_framework(system_message, user_message)
```
