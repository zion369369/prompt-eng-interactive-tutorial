## 📦 OpenAI JSON Mode
When using `response_format={ "type": "json_object" }`:
1. You must explicitly instruct the model to return JSON in the system or user prompt.
2. The model will output a valid JSON string that can be loaded directly with `json.loads()`.

Let's see how to configure it.


```python
from openai import OpenAI

# Example configuration
messages = [
    {
        "role": "system",
        "content": """
        You are an operations assistant.
        Extract the key action items and deadlines from the message and return them as a JSON object.
        The JSON must contain a single key 'action_items', containing a list of tasks and assignees.
        """
    },
    {
        "role": "user",
        "content": "Jane must finalize graphics by Wednesday. Bob needs to write release notes by Friday."
    }
]

# To call the API with JSON mode enabled:
# client = OpenAI()
# response = client.chat.completions.create(
#     model="gpt-5.4-mini",
#     response_format={"type": "json_object"},
#     messages=messages
# )
```

## 🏋️ Interactive Exercise: Vendor Proposal Comparison Schema
Write a prompt instruction that will return a JSON object comparing two proposals (e.g. Spa Cleaning Proposals).
Specify that the JSON object must contain keys `vendor_a_cost`, `vendor_b_cost`, and `comparison_notes`.

Write your JSON instruction text below.


```python
# Define the prompt instruction containing your target keys:
json_prompt_instruction = """
[Instruct the model to return a JSON object with vendor_a_cost, vendor_b_cost, and comparison_notes]
"""
```

```python
# Validator to check instruction requirements
def validate_json_prompt(p):
    if "[Instruct" in p or not p.strip():
        print("❌ Please write your prompt instruction first!")
        return
        
    score = 0
    criteria = []
    
    required_keys = ["vendor_a_cost", "vendor_b_cost", "comparison_notes", "json"]
    found = [k for k in required_keys if k in p.lower()]
    
    if len(found) >= 3:
        score += 1
        criteria.append(f"✅ Prompt requests all required keys: {found}")
    else:
        criteria.append(f"⚠️ Prompt is missing key details (required keywords: {required_keys})")
        
    print(f"Validation score: {score}/1")
    for c in criteria:
        print(c)
        
    if score == 1:
        print("\n🎉 Excellent! This ensures the JSON parser receives the expected structure.")
    else:
        print("\n💡 Revise the instructions to ensure all keys are mentioned.")

validate_json_prompt(json_prompt_instruction)
```
